```python
import pandas as pd
import streamlit as st
import numpy as np
import altair as alt
import json
```

```python
census_file = "DDW_PCA0000_2011_Indiastatedist.xlsx"
```

```python
df = pd.read_excel(census_file)
```

```python
df.tail()
```

```python
df.columns.values
```

```python
states_df = df.loc[(df["Level"] == "STATE") & (df["TRU"] == "Total") ]
```

```python
filtered_states_df = states_df[["State","Name","TOT_P","TOT_M","TOT_F"]]
filtered_states_df
```

```python
filtered_states_df[["State","Name"]]
```

```python
states_name_and_code_file = "Indian_States_TIN_Codes.xlsx"
```

```python
state_name_code_df = pd.read_excel(states_name_and_code_file)
state_name_code_df.sort_values(by="TIN Code")
```

```python
merged_column = filtered_states_df.merge(right=state_name_code_df[["TIN Code","2-Letter Code"]], left_on="State",right_on="TIN Code")
merged_column.drop(columns=["State"])
```

```python
india_topo_json_file_name = "india.json"
```

```python
topo_data = alt.topo_feature(india_topo_json_file_name,"states")
```

```python
lookup_data = alt.LookupData(merged_column,'State',['TOT_P'])
```

```python
alt.Chart(topo_data).mark_geoshape().encode(
    color='Properties.st_code:Q'
).transform_lookup(
    lookup='id',
    from_=lookup_data
).properties(
    width=500,
    height=300
).project(type='mercator')
```
